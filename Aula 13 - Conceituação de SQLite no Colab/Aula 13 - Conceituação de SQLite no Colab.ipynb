> A query abaixo retorna a m√©dia de unidades vendiddas por produto, se o valor unit√°rio for maior do que 199, e somente se a m√©dia de unidades vendidas for maior do que 10

```python
query5 = '''SELECT Nome_Produto, AVG(Unidades_Vendidas) FROM tb_vendas_dsa 
            WHERE Valor_Unitario > 199 GROUP BY Nome_Produto HAVING AVG(Unidades_Vendidas) > 10'''
```

### Executa a query no BD

```python
cursor.execute(query5)
```

### Visualiza resultado

```python
cursor.fetchall()
```

# Aplicando Linguagem SQL na Sintaxe do Pandas com Linguagem Python

### Estabelece a conex√£o

```python
con = sqlite3.connect('db_aula.db')
```

### Abre um cursor para percorrer os dados do BD

```python
cursor = con.cursor()
```

## A query abaixo retorna todas as linhas e todas as colunas da tabela

```python
query = 'SELECT * FROM tb_vendas_dsa'
```

> Executa a query

```python
cursor.execute(query)
```

```python
dados = cursor.fetchall()
```

```python
dados
```

```python
type(dados)
```

## Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame(dados, columns= ['ID_Pedido', 'ID_Cliente', 'Nome_Produto', 'Valor_Unitario', 
                                   'Unidades_Vendidas', 'Custo'])
```

```python
df.head()
```

### Fechando o cursor e encerrando a conex√£o

```python
cursor.close()
con.close()
```

> Retornar m√©dia de unidades vendidas 

```python
media_unidades_vendidas = df['Unidades_Vendidas'].mean()
```

```python
type(media_unidades_vendidas)
```

```python
print(media_unidades_vendidas)
```

> Retornar m√©dia de unidades vendidas por produto 

```python
media_unidades_vendidas_por_produto = df.groupby('Nome_Produto')['Unidades_Vendidas'].mean()
```

```python
media_unidades_vendidas_por_produto.head(10)
```

> Retornar a m√©dia de unidades vendiddas por produto, se o valor unit√°rio for maior do que 199, e somente se a m√©dia de unidades vendidas for maior do que 10

## Sintaxe Pandas

```python
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto').filter(lambda x: x['Unidades_Vendidas'].mean() > 10)
```

## Sintaxe SQL

```python
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto') \
                              .filter(lambda x: x['Unidades_Vendidas'].mean() > 10) \
                              .groupby('Nome_Produto')['Unidades_Vendidas'].mean()
```
